```python
from transformers import pipeline
```

```python
classifier=pipeline("sentiment-analysis")
```

```python
classifier("This library is good enough")
```

```python
classifier("But these toilet are not too much hygeinic ")
```

```python
sents=["The location is great","Area is not so good","Facility are good enough","AC are not too much cooling"]
result=classifier(sents)
```

```python
for i,j in zip(sents,result):
  print(f"Sentence is: {i} and Result of this Sentence is :{j}")
```

```python
from transformers import AutoTokenizer, TFAutoModelForSequenceClassification
#these two work as the blueprint of the model
#1. First one is used for the tokenize the sentences with respect to which model is chosen and for the different-different purpose the model
#Would also be changes countinuously.....
```

```python
model_name="nlptown/bert-base-multilingual-uncased-sentiment"  #chose the which model you want for which type of task.
```

```python
#this model is available on the pytorch so we use "from_pt" to import this model in the tensorflow
model=TFAutoModelForSequenceClassification.from_pretrained(model_name,from_pt=True)  #we are training model which we want from the model we select for our purpose
tokenizer=AutoTokenizer.from_pretrained(model_name)  #we are training the tokenizer according to us which we model for which purpose we select
classifier_model=pipeline("sentiment-analysis",model=model,tokenizer=tokenizer)   #now give the model and tokenizer according to your model which you select.
```

```python
classifier_model("Hey! This is very beautiful sights from your side")
```
